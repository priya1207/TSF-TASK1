#STEP 1- IMPORT THE LIBRARIES AND DATASET 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
from plotnine import *
import warnings
warnings.filterwarnings('ignore')
```

```python
sample = pd.read_csv('/content/SampleSuperstore.csv')
sample.head(15)
```

#STEP-2 VISULAIZING THE DATA

```python

plt.figure(figsize=(20,10))
plt.plot('Sub-Category','Category', data=sample)
plt.title('Category vs Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=50)
plt.show()
```

```python
Profit_plot=(ggplot(sample, aes(x='Sub-Category', y='Profit', fill='Sub-Category')) + geom_col() + coord_flip()
+ scale_fill_brewer(type='div', pelette='Spectral') + theme_classic() + ggtitle('pie chart'))
display(Profit_plot)
```
